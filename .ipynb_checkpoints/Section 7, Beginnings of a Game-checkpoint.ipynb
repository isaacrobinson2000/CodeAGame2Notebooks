```python
%%javascript

function game(timeStep, gameState) {
    let painter = gameState.painter;
    let {width, height} = gameState.canvas;
    
    painter.clearRect(0, 0, width, height);
    painter.fillStyle = "black";
    painter.font = "30px Ariel";
    
    let s = "";
    
    for(let k in gameState.keysPressed) {
        s += k + " ";
    }
    
    painter.fillText("Keys Pressed: " + s, 0, 30);
}

element.makeGame(game);
```
